```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
credits.head(1)
```

```python
#Merging the data fram
movies = movies.merge(credits,on='title')
```

```python
movies.head(1)
```

```python
#Dataset Cleaning based on Content
#1.genres
#2.id
#3.keywords
#4.title
#5.overview
#6.cast
#7.crew

movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
#Data preprocessing
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
#'[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]'
#We have to Convert above form into below dict
#['Action','Adventure','Fantasy','SciFi']
```

```python
# Creating a function
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L    
```

```python
import ast
#ast.literal_eval()
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
#Same thing we have to do for keywords
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
#Creating a function for cast to get the first three important name of the character
def convert3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter !=3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L        
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
#Creating Function for crew
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L         
    
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
movies['overview'][0]
```

```python
#Coverting the Overview into list so that at the end we can add to list
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
#Removing the space between the names
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
# Creating a Column for a tag which will be the copncatenation of overview,genres,keywords,cast,crew
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
# Creating a New dataframe where we take only three things movie_id,title,tags
new_dataframe = movies[['movie_id','title','tags']]
```

```python
new_dataframe
```

```python
# Converting the tags list into string
new_dataframe['tags'] = new_dataframe['tags'].apply(lambda x:" ".join(x))
```

```python
new_dataframe.head()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)    
        
```

```python
new_dataframe['tags'] = new_dataframe['tags'].apply(stem)
```

```python
new_dataframe['tags'][0]
```

```python
new_dataframe['tags'] = new_dataframe['tags'].apply(lambda x:x.lower())
```

```python
new_dataframe.head()
```

```python
# TExt Vectorization
# We have to convert the text to vector 
#Problem-calculating the similarity between the two tags
# Technique used is "Bag of words"
# cv is the object
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(new_dataframe['tags']).toarray()
```

```python
vectors
```

```python
cv.get_feature_names()
```

```python
ps.stem('loved')
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_similarity(vectors)
```

```python
similarity = cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
#Creating a function for recommend
def recommend(movies):
    movie_index = new_dataframe[new_dataframe['title'] == movies].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(new_dataframe.iloc[i[0]].title)
        
    
```

```python
recommend('Avatar')
```
